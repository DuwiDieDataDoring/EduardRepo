```
#Usual Suspects
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import nltk
from sklearn.preprocessing import StandardScaler,RobustScaler
from sklearn.pipeline import make_pipeline
```

```
train = pd.read_csv('../training.csv')
test = pd.read_csv('../test.csv')
```

```
train
```

```
train.corr()
```

```
train_features = train.drop('FraudResult', axis = 1)
y = train['FraudResult'].reset_index(drop=True)
test_features = test
features = pd.concat([train_features, test_features])
```

```
def remove_vocab(post):
    return ''.join(i for i in post if i.isdigit())

features['TransactionId'] = features['TransactionId'].apply(remove_vocab)
features['BatchId'] = features['BatchId'].apply(remove_vocab)
features['AccountId'] = features['AccountId'].apply(remove_vocab)
features['SubscriptionId'] = features['SubscriptionId'].apply(remove_vocab)
features['CustomerId'] = features['CustomerId'].apply(remove_vocab)
features['ProviderId'] = features['ProviderId'].apply(remove_vocab)
features['ProductId'] = features['ProductId'].apply(remove_vocab)
features['ChannelId'] = features['ChannelId'].apply(remove_vocab)

features['TransactionId'] = features['TransactionId'].astype(int)
features['BatchId']= features['BatchId'].astype(int)
features['AccountId'] = features['AccountId'].astype(int)
features['SubscriptionId'] = features['SubscriptionId'].astype(int)
features['CustomerId'] = features['CustomerId'].astype(int)
features['ProviderId'] = features['ProviderId'].astype(int)
features['ProductId'] = features['ProductId'].astype(int)
features['ChannelId'] = features['ChannelId'].astype(int)

```

```
features
```

```
features = features.drop(['BatchId','AccountId','SubscriptionId','CustomerId','CurrencyCode','CountryCode','PricingStrategy','TransactionStartTime','ProductId','ProviderId','ChannelId'], axis = 1)
```

```
features
```

```
final_features = pd.get_dummies(features)
```

```
final_features
```

```
X = final_features[:len(y)]
test_X = final_features[len(X):]
X.head()
```

```
X.drop(['TransactionId'], axis=1, inplace=True)
test_X.drop(['TransactionId'], axis=1, inplace=True)
```

```
X_train = X
X_test = test_X
y_train = y
```

```

model_class = RandomForestClassifier()

model = make_pipeline(RobustScaler(), model_class)

model.fit(X_train, y_train)
```

```
pred = model.predict(X_test)
pred
```

```
submission = pd.DataFrame({'TransactionId':test['TransactionId'], 'Fraudulent':pred})
submission.to_csv('submission.csv', index=False)
```
